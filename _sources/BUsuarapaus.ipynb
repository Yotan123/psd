1.  Latar Belakang Masalah

Paus kanan Atlantik Utara (North Atlantic Right Whale) merupakan salah satu spesies paus paling terancam punah di dunia. Populasinya diperkirakan kurang dari 400 individu dan terus menurun akibat tabrakan kapal, aktivitas industri laut, serta perubahan lingkungan. Oleh karena itu, pemantauan populasi paus ini menjadi sangat penting bagi para peneliti dan konservasionis.

Salah satu metode utama untuk memantau keberadaan paus adalah analisis suara bawah laut (bioacoustic). Paus kanan menghasilkan suara khas seperti up-calls, yaitu suara frekuensi rendah yang digunakan untuk komunikasi. Dengan mengklasifikasikan pola suara tersebut, kita dapat mendeteksi keberadaan paus secara otomatis tanpa harus melakukan observasi manual.

2. Permasalahan Bisnis

Masalah utama yang ingin dipecahkan adalah bagaimana mendeteksi keberadaan paus kanan secara otomatis dari data audio laut. Data rekaman audio bawah laut biasanya:
   
*   Panjang durasi rekaman besar,
*   Banyak noise (kapal, gelombang, hewan lain),
*   Tidak mungkin dianalisis satu per satu secara manual.

Tanpa sistem otomatis, proses identifikasi paus akan:
* Memakan waktu lama,
* Membutuhkan tenaga ahli akustik laut,
* Tidak efisien untuk skala besar.

3. Tujuan Bisnis(Tujuan Analisis)

Tujuan dari proyek ini adalah membangun model klasifikasi time series yang mampu:

* Membedakan apakah suatu rekaman audio mengandung suara paus kanan atau tidak,

* Mengurangi beban kerja manual dalam identifikasi suara,

* Mendukung pemantauan populasi paus secara lebih cepat dan akurat.

Dengan model klasifikasi yang baik, organisasi konservasi laut dapat:

* Menentukan area migrasi paus dengan lebih akurat,

* Mengurangi resiko tabrakan kapal,

* Merancang kebijakan perlindungan habitat secara lebih efektif.

4. Ruang Lingkup Proyek

Dalam proyek klasifikasi time series ini, scope utamanya meliputi:

* Pengolahan dataset RightWhaleCalls,

* Ekstraksi fitur atau penggunaan sinyal raw,

* Pelatihan model klasifikasi time series,

* Evaluasi menggunakan akurasi, precision, recall, dan f1-score,

* Visualisasi performa model dan interpretasi hasil.

5. Dampak Jangka Panjang

Model deteksi otomatis dapat diterapkan pada sistem pemantauan laut real-time (misalnya buoy sensor), sehingga upaya konservasi paus lebih adaptif dan responsif terhadap perubahan.
